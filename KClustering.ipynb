```python
data=df.drop('Country',axis=1)
```

```python
data
```

```python
X=pd.get_dummies(data=data)
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
ssd=[]
OMP_NUM_THREADS=1
for k in range(2,30):
    model=KMeans(n_clusters=k)
    model.fit(X_scaled)
    ssd.append(model.inertia_)
```

```python
plt.plot(np.arange(2,30),ssd,'o--')
```

```python
tmp=pd.Series(ssd).diff()
```

```python
plt.figure(dpi=150)
plt.plot(tmp,'o--')
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```

```python
#K=3
model=KMeans(n_clusters=3)
clusters=model.fit(X_scaled)
```

```python
X['3 Clusters']=model.labels_
```

```python
t=X.corr()['3 Clusters'].sort_values().drop('3 Clusters',axis=0)
X.corr()['3 Clusters'].sort_values().drop('3 Clusters',axis=0)
```
