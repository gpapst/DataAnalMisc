```python
X=df.drop(['Churn','customerID'],axis=1)
X=pd.get_dummies(X,drop_first=True)
```

```python
y=df['Churn'].apply(lambda x: 1 if x=="Yes" else 0)
y
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
```

```python
model1=DecisionTreeClassifier()
```

```python
grid1=GridSearchCV(estimator=model1,param_grid={'max_depth':[2,4,5,6],'criterion':['gini',"entropy", "log_loss"]})
```

```python
grid1.fit(X=X_train,y=y_train)
grid1.best_params_
```

```python
preds=grid1.best_estimator_.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_true=y_test,y_pred=preds)
```

```python
features = pd.DataFrame(data=grid1.best_estimator_.feature_importances_,index=X.columns,columns=['Importance'])
features.sort_values('Importance')
```

```python
from sklearn.tree import plot_tree
plt.figure(dpi=300)
plot_tree(grid1.best_estimator_,filled=True,feature_names=X.columns);
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier
```

```python
model2=RandomForestClassifier()
```

```python
model2.fit(X_train,y_train)
```

```python
preds2=model2.predict(X_test)
```

```python
print(classification_report(y_true=y_test,y_pred=preds2))
```
